```python
import pandas as pd

gamelogs = pd.read_csv('Gamelogs19-20.csv')

gamelogs.rename(columns={'Rk ATL':'Team'}, inplace=True)

gamelogs
```

```python
gamelogs2 = pd.read_csv('Gamelogs19-20.csv')

gamelogs2.rename(columns={'Opp':'Team'}, inplace=True)

gamelogs2
```

```python
# make a column with the index values
gamelogs2.reset_index(inplace=True)

# merge the two dataframes based on the selected columns.
combined = gamelogs.merge(gamelogs2, on=['Date', 'Team'])

combined
```

```python
combined = combined.iloc[:,0:40]

combined.rename(columns={'DRtg_y':'OppDefRtg'}, inplace=True)

combined
```
